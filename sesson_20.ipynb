```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(-10,9,20)
b = np.linspace(-10,9,20)
a,b
```

```python
xx , yy = np.meshgrid(a,b)
yy,xx
```

```python
plt.scatter(xx,yy)
```

```python
def func(x,y):
  return 2*(x)+ 2*y

zz = func(xx,yy)
zz
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = px.scatter_3d()
fig.add_trace(go.Surface(x =xx , y = yy ,z=zz))
fig.show()
```
